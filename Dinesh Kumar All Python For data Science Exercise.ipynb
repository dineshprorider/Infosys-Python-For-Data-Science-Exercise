# Numpy Excercise 

```python
import numpy as np
data = np.array([
    [1, 6012],
    [2, 7079],
    [3, 6886],
    [4, 7230],
    [5, 4598],
    [6, 5564],
    [7, 6971],
    [8, 7763],
    [9, 8032],
    [10, 9569]
])
data[:, 1] += 2000
print(data)
```

```python
steps_more_than_9000 = data[data[:, 1] > 9000][:, 1]
print("Steps walked more than 9000:", steps_more_than_9000)
```

```python
sorted_steps = np.sort(data[:, 1])
print("Steps walked in sorted order:", sorted_steps)
```

# Pandas Excercise

```python
import pandas as pd
data = pd.read_csv("rainfall1627650742214.csv")
data
```

```python
print("Missing values per column:\n", data.isnull().sum())
```

```python
data = data.dropna()
data
```

```python
highest_rainfall_district = data.loc[data['ANNUAL'].idxmax()]
print("District with the highest annual rainfall:\n", highest_rainfall_district[['DISTRICT', 'ANNUAL']])
```

```python
state_annual_rainfall = data.groupby('STATE_UT_NAME')['ANNUAL'].sum()

top_5_states = state_annual_rainfall.nlargest(5)
print("Top 5 states with the highest annual rainfall:\n", top_5_states)
```

```python
data = data.drop(columns=['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'])
print("Data after dropping specified columns:\n", data.head())
```

```python
pivot_table = data.pivot_table(index='STATE_UT_NAME', values=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'], aggfunc='mean')
print("State-wise mean rainfall for all the months:\n", pivot_table)
```

```python
district_count_per_state = data['STATE_UT_NAME'].value_counts()
print("Count of districts in each state:\n", district_count_per_state)
```

```python
max_rainfall_districts_may = data.loc[data.groupby('STATE_UT_NAME')['MAY'].idxmax()][['STATE_UT_NAME', 'DISTRICT', 'MAY']]
print("For each state, the district that gets the highest rainfall in May and the recorded rainfall:\n", max_rainfall_districts_may)
```

# Matplotlib Exercise

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2*np.pi*x) * np.exp(-x)

plt.plot(x, y, marker='o', linestyle='-')

plt.xlabel('Abscissa')
plt.ylabel('Ordinate')
plt.title('Main heading\nFigure Title')

plt.grid(True)

plt.xlim(0, 5)
plt.ylim(-1, 1.2)

plt.xticks(np.arange(0.0, 5.1, 1.0))
plt.yticks(np.arange(-1.0, 1.3, 0.2))

plt.show()
```
